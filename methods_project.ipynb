## 1. Resumo

A Ararinha-Azul (Cyanopsitta spixii) é uma espécie de ave psitacídea endêmica do Brasil, cujo habitat se limita a uma pequena área da caatinga nordestina no extremo norte da Bahia, ao sul do Rio São Francisco. Declarada oficialmente extinta da natureza em 2000, grandes esforços têm, desde então, sido movidos internacionalmente para repatriar os indivíduos remanescentes em cativeiro e reintroduzir a espécie em seu habitat natural.

Apesar da mobilização internacional de órgãos e pesquisadores para a recuperação da espécie, a falta de dados comportamentais registrados da espécie e a quantidade escassa de indivíduos remanescentes limitam a execução do programa e aumentam a possibilidade de falha nas tentativas, fazendo-se necessária uma abordagem metódica com margem de erro ínfima.

Neste trabalho, objetiva-se a apresentação de um modelo preciso que permita simular o desenvolvimento em função do tempo das comunidades de Ararinhas-Azuis em um cenário próximo ao observado na natureza. Nesse sentido, almeja-se a possibilidade de análise das melhores condições ecológicas e ambientais para a sobrevivência e manutenção da população reintroduzida, sem o comprometimento da integridade dos indivíduos remanescentes.


## 2. Introdução e Objetivos



A espécie foi descoberta em 1819 por Johann Baptist Ritter von Spix (em Juazeiro, Bahia), mas só veio a ser efetivamente descrita em 1832 por Johann Waggler. Permanecendo, então, sem registros até sua ocasional redescoberta por Paul Roth, em 1986, quando foram encontrados os três últimos exemplares selvagens da espécie no município de Curaçá, no nordeste da Bahia.
Até então, sua região de ocorrência permanecera desconhecida e sua distribuição geográfica histórica ainda é indefinida, mas a redescoberta das aves por parte de Roth tornou possível determinar a área de abrangência da espécie.

Em 1990, durante uma expedição liderada por Carlos Yamashita e Tony Jupiter, foi identificado o único indivíduo remanescente na natureza (um macho). Foi iniciado, então, um projeto para localização de outros indivíduos e recuperação da espécie, mas nenhuma investida apresentou sucesso significativo. Com a morte do macho selvagem, registrada em 2000, e a posterior não identificação de novos indivíduos, a espécie foi declarada extinta na natureza.

Apesar de ocasionais relatos de avistamentos por moradores locais, nenhum registro oficial de indivíduos da espécie foi documentado durante duas décadas. Em 2019, havia apenas 166 aves registradas em cativeiro, 13 delas no Brasil e as demais distribuídas pela Alemanha (147), Bélgica (4) e Singapura (2). Um novo projeto de reinserção da espécie na natureza foi iniciado em 2020 com a repatriação de 52 Ararinhas-Azuis originadas no criadouro alemão. No momento da chegada das aves em território brasileiro (03/03/2020), existiam apenas 180 exemplares em cativeiro (22 no Brasil, 152 na Alemanha, 4 na Bélgica e 2 em Singapura).

Em 2022, foram liberados dois grupos de indivíduos que estavam passando por treinamento para se adaptarem à vida em liberdade: o primeiro grupo com 8 aves em 11 de junho, e o segundo com 12 aves em 10 de dezembro. No entanto, o projeto já registra algumas perdas, pois das 20 Ararinhas soltas, três foram alvo de predação por aves de rapina da região e uma está desaparecida.

Dado o estado crítico atual em que se encontra a espécie, o sucesso dessa iniciativa pode determinar o seu retorno à paisagem da caatinga baiana ou decretar sua extinção definitiva. Tendo isso em vista, pontos de interesse podem ser definidos para a análise dos modelos:

1. **Desenvolvimento da população ativa:** crescimento da população de Ararinhas-Azuis adultas, possibilitando a reprodução entre os indivíduos e o crescimento da população de Ararinhas-Azuis jovens.
2. **Desenvolvimento da população inativa:** crescimento da população de Ararinhas-Azuis jovens, possibilitando a maturação dos indivíduos e o crescimento da população de Ararinhas-Azuis adultas.
3. **Desenvolvimento sustentável da espécie:** conservação temporal das populações de Ararinhas-Azuis jovens e adultas, atingindo uma estabilidade populacional e permitindo a manutenção da espécie na natureza.
4. **Decaimento da população ativa:** redução da população de Ararinhas-Azuis adultas, limitando a reprodução entre os indivíduos e o crescimento da população de Ararinhas-Azuis jovens.
5. **Decaimento da população inativa:** redução da população de Ararinhas-Azuis jovens, limitando a maturação dos indivíduos e o crescimento da população de Ararinhas-Azuis adultas.
6. **Extinção:** redução da população de Ararinhas-Azuis jovens e/ou adultas a níveis críticos que incapacitam a capacidade de manutenção da espécie, eliminando ambas as populações.

Nesse sentido, o modelo matemático apresentado tem como objetivo simular as dinâmicas populacionais das Ararinhas-Azuis inseridas em um cenário com a influência de agentes antagônicos representados pelos predadores de ninho e aves de rapina, com o intuito de prever de forma confiável o desenvolvimento da população em ambiente natural e definir os cenários de maior êxito da espécie.

## 3. Modelo




O modelo utilizado como base do projeto foi extraído do artigo “Simulating the Reintroduction of the Extinct Passenger Pigeon (Ectopistes migratorius)”[1], o qual foi proposto e desenvolvido por  Erin Boggess, Jordan Collignon e Alanna Riederer, o qual aborda um modelo de equações diferenciais que simula a dinâmica populacinal de uma espécie extinta de pombos nativa dos Estados Unidos, o pombo-passageiro, após sua reintrudução em um ambiente natural com a influência de predadores. As  equações diferenciais foram definidas utilizando os modelos existentes no artigo e baseando-se para contemplar a relação ecológica entre populações de Ararinha-Azul e predadores. Segue modelo de equação diferencial:


$I. \dfrac{dY}{dt} =  αA(1-\dfrac{A}{K_1}) (\dfrac{A}{ϵ} - 1)- μY - N (\dfrac{β_1Y^2}{1+β_1ψ_1Y^2}) - π_1Y $ , \
$II.\dfrac{dA}{dt} =  µY − R (\dfrac{β_2A^2}{1+β_2ψ_2YA^2}) - π_2A$ , \
$III.\dfrac{dN}{dt} = λ_1N (1-\dfrac{N}{K_2}) + γ_1Y N$, \
$IV.\dfrac{dR}{dt} = λ_2R (1-\dfrac{R}{K_3}) + γ_2RA$

### Especificação do modelo


O sistema é baseado no modelo clássico de Lotka-Volterra que tem o intuito de capturar a dinâmica presa-predador entre duas espécies. Aqui, ele é aplicado para dois sistemas isolados: Ararinhas adultas (A) e aves de rapina (R), e Ararinhas jovens (Y) e predadores de ninho (N). O modelo de equações diferenciais de Lotka-Volterra é definido a seguir:

$I. \dfrac{dP_1}{dt} = αP_1-βP_2$ \\
$II. \dfrac{dP_2}{dt} = -γP_2+vP_1P_2$

Além do modelo de Lotka-Volterra, também foi incorporado ao sistema o modelo de crescimento populacional logístico para simular um crescimento populacional mais realístico. Os modelos de crescimento logístico são determinados pela razão entre a população e a capacidade de carga da mesma, que expressa a quantidade de indivíduos que o ambiente é capaz de suportar com seus recursos sem gerar degradação do meio ou impactos negativos para a população. O modelo de crescimento logístico está representado abaixo:

$I. \dfrac{dP_1}{dt} = αP_1\left(1-\dfrac{P_1}{K_1}\right)-βP_2$ \
$II. \dfrac{dP_2}{dt} = -γP_2+vP_1P_2\left(1-\dfrac{P_2}{K_2}\right)$

O próximo passo de desenvolvimento do modelo foi a separação dos sistemas de presa-predador baseado na descrição de diferentes tipos de predadores. Esses foram divididos em dois grupos maiores com base no seu comportamento de predação: serpentes (como a jararaca), pequenos primatas (como o sagui) e aves de médio porte (como o tucano) predam os ovos e os indivíduos jovens, e aves de rapina (como o gavião-pernilongo) predam indivíduos adultos.

Considerando essa divisão de predadores, também foi feita uma divisão da população de Ararinhas em subpopulações de indivíduos adultos e juvenis com influências de grupos de predadores distintos. Nesse sentido, é importante considerar para análise do modelo o processo de maturação dos pombos jovens em indivíduos adultos. O modelo com equações separadas por influência dos predadores é muito semelhante ao resultado final usado para modelagem do cenário. A seguir estão apresentadas as equações do sistema:

$I. \dfrac{dY}{dt} =  αY\left(1-\dfrac{Y}{K_1}\right) - μY  - β_1N,$ \
$II.\dfrac{dA}{dt} = αA\left(1-\dfrac{A}{K_2}\right) + µY - β_2R$ , \
$III.\dfrac{dN}{dt} = -γ_1N+v_1YN\left(1-\dfrac{N}{K_3}\right)$, \
$IV.\dfrac{dR}{dt} = -γ_2R+v_2AR\left(1-\dfrac{R}{K_4}\right)$

Por fim, a última etapa de formulação do modelo foi incluir o efeito de Allee e o Holling type para as relações de predação nas considerações de dinâmica populacional das Ararinhas. O efeito Allee é um fenômeno que expressa a taxa de crescimento negativa de populações pequenas ou com baixa densidade. Enquanto o Holling type é um parâmetro que descreve o comprtamento de adaptação dos predadores à presença de presas novas no ambiente, simulando o processo de aprendizado por parte dos mesmos. Essas abordagens serão mais amplamente discutidas em uma etapa posterior quando se iniciar a discussão dos parâmetros utilizados no sistema. O modelo completo incluindo o efeito Allee e o Holling type está retratado abaixo:

$I. \dfrac{dY}{dt} =  αA\left(1-\dfrac{A}{K_1}\right) \left(\dfrac{A}{ϵ} - 1\right)- μY - N \left(\dfrac{β_1Y^2}{1+β_1ψ_1Y^2}\right) - π_1Y $ , \
$II.\dfrac{dA}{dt} =  µY - R \left(\dfrac{β_2A^2}{1+β_2ψ_2YA^2}\right) - π_2A$ , \
$III.\dfrac{dN}{dt} = λ_1N \left(1-\dfrac{N}{K_2}\right) + γ_1Y N$, \
$IV.\dfrac{dR}{dt} = λ_2R \left(1-\dfrac{R}{K_3}\right) + γ_2RA$

### 3.1. Delay Differential Equation (DDE)

Além do modelo de ODEs, também cabe trazer para discussão a aplicação, no modelo, de DDEs similares, com a inclusão de um parâmetro de atraso de tempo $τ$. Ao modelar dinâmicas populacionais, o parâmetro de atraso é empregado com o intuito de simular o atraso no crescimento populacional ocasionado pelo tempo de gestacão e, no modelo que estamos tratando, o tempo de incubação das Ararinhas se torna um fator comparável, servindo como o atraso para as DDEs. Como a incubação é relacionada com a população das Ararinhas, o parâmetro é aplicado como atraso nas equações da população de Ararinhas jovens e adultas (equações I e II, respectivamente), dando um maior teor de realismo e complexidade para o modelo. O sistema empregando DDEs para considerar o tempo de incubação dos filhotes é apresentado abaixo:


$I. \dfrac{dY}{dt} =  αA(1-\dfrac{A}{K_1}) (\dfrac{A}{ϵ} - 1)- μY(t - τ) - N (\dfrac{β_1Y^2}{1+β_1ψ_1Y^2}) - π_1Y $ , \
$II.\dfrac{dA}{dt} =  µY(t - τ) - R (\dfrac{β_2A^2}{1+β_2ψ_2YA^2}) - π_2A$




Apesar da implementação de sistemas de DDE não ser complexa para modelagem de um cenário de interação entre populações, o sistema expressa uma abordagem mais determinista, apresentando uma maior exigência de valores apropriados para a definição dos parâmetros utilizados. No entanto, a ausência de dados relacionados ao tópico de discussão impossibilita a determinação de valores adequados para os parâmetros. Ressalta-se, ainda, a importância de manter uma variabilidade estocástica no modelo por tratar-se da modelagem de um cenário de reintrodução da espécie. Dado a esses fatores, a análise do sistema de DDE não foi considerada para as etapas posteriores do projeto, uma vez que não convém com os objetivos discutidos.



### 3.2. Inicialização do Modelo de ODE para Alimento e Presa

Ainda, dentro do escopo de discussão deste trabalho, outro fator essencial a se trazer para análise a fim de atingir uma melhor aproximação de um cenário natural real é a disponibilidade de alimento no ambiente da reintrodução e o impacto na taxa de crescimento da população de Ararinhas decorrente do consumo desse alimento. Para consideração dessas variáveis no sistema, será empregado o modelo extraído do artigo "rescomp: An R package for efficient ODE modelling of resource competition"[11] que determina o impacto da interação entre populações de consumidores e os suprimentos de recursos na densidade populacional e concentração de recursos no ambiente. O modelo de interação consumidor-recurso é apresentado a seguir:

 \begin{equation}
 \dfrac{dN_i}{dt}=N_i(μ_i(R_1,R_2,...)-m) \\
 \dfrac{dR_j}{dt}=Ψ_j(R_j)-\sum_{i=1}^{n}Q_{ij}μ_{ij}(R_j)N_i
 \end{equation}

No cenário analisado, serão tratados apenas uma população de consumidores — as Ararinhas adultas (A) — e um suprimento de recursos — as principais fontes de alimento das Ararinhas: o pinhão-bravo (*Jatropha mollissima*) e a faveleira (*Cnidoscolus quercifolius*). Para evitar confusões com a já existente classe $R$ que representa as aves de rapina, a equação de descrição dos recursos será chamada de $F$. O modelo completo finalizado está descrito abaixo:

 \begin{equation}
I. \quad \frac{dY}{dt} =  αA\left(1-\frac{A}{K_1}\right) \left(\frac{A}{ϵ} - 1\right) - μY - N \left(\frac{β_1Y^2}{1+β_1ψ_1Y^2}\right) - π_1Y , \\
II. \quad \frac{dA}{dt} =  µY + \kappa AF - R \left(\frac{β_2A^2}{1+β_2ψ_2YA^2}\right) - π_2A , \\
III. \quad \frac{dN}{dt} = λ_1N \left(1-\frac{N}{K_2}\right) + γ_1Y N , \\
IV. \quad \frac{dR}{dt} = λ_2R \left(1-\frac{R}{K_3}\right) + γ_2RA , \\
V. \quad \frac{dF}{dt} = rF\left(1-\frac{F}{K_4}\right)-\kappa QFA
\end{equation}


```python
# k é a constante de consumo do recurso
# r é a taxa de "produção" do recurso
# Q é a cota de consumo do recurso (quantidade consumida por indivíduo)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model_2(y, t, a, u, p1, p2, l1, l2, b1, b2, b3, y1, y2, w1, w2, K1, K2, K3, k4, e, q, k, r):
  Y, A, N, R, F = y
  sol_system = [a*A*(1-(A/K1))*((A/e)-1) - u*Y - N*(b1*Y**2/(1+b1*w1*Y**2)) - p1*Y,
                u*Y + (b3*(F/(k+F)))*A - R*(b2*A**2/(1+b2*w2*A**2)) - p2*A,
                l1*N*(1-(N/K2)) + y1*Y*N,
                l2*R*(1-(R/K3)) + y2*R*A,
                r*F*(1-(F/K4)) - q*A*(b3*(F/(k+F)))]
  return sol_system
```

```python
#------- condições iniciais --------

Y0 = 2 # starting young pigeons
A0 = 2 # starting adult pigeons
N0 = 2 # starting nest predators
R0 = 2 # starting raptors (predators of adult pigeons)
F0 = 3 # Quatidae de recurso no ambiente
y0 = Y0, A0, N0, R0, F0
t = np.linspace(0, 150, 150) #análise do experiemento nos primeiros anos de reinserção (consideraremos o modelo em meses)

#------- constantes --------
Fmax = 9
a = 0.08 # birth of new pigeons / definido a partir de dado particular do comportamento de uma comunidade de 52 Ararinhas em cativeiro
m = 0.6 # maturation of young pigeons into adult pigeons (tempo de virarem adultos eh 6 meses)
p1 = 0.30 # death of young pigeons (Dado empírico, com valor mais alto considerando a fragilidade de filhotes de aves)
p2 = 0.15 # death of adult pigeons (Dado empírico, capacidade de morte de um adulto em porcentagem)
l1 = 0.09 # growth of nest predators (Dado empírico, consideramos baixo dado que o índice de nascimento de filhotes é menor do que a de total de adultos Ararinha)
l2 = 0.1 # growth of raptors
## Holling Type parameters: taxa populacional do predador em função da densidade da presa.
## Type I: taxa de crescimento linear; assume que o consumo e digestão do alimento não interferem na sua busca.
## Type II: taxa de crescimento hiperbólica (aR/(1+ahR); assume que a busca por alimento é limitada pelo seu processamento.
## Type III: taxa de crescimento hiperbólica potencial (aR^k/(1+ahR^k)); simula processo de aprendizado do predador.
b1 = 0.07 # Holling Type parameter for the young pigeons/nest predators system
b2 = 0.096 # Holling Type parameter for the adult pigeons/raptors system
b3 = 3 #Holling Type Parameter tipo 2 da ararinha em função do aliemnto (aR/(1+ahR)
g1 = 0.01 # predation for the nest predators (Dado empirico,cada predado come 1)
g2 = 0.01 # predation for the raptors (1 a cada)
ps1 = 4 # handling time for the young pigeons (Dado empírico, estimativa de tempo (dado em horas))
ps2 = 4 # handling time for the adult pigeons (Dado empírico, estimativa de tempo (dado em horas))
K1 = b3 * F0 # carrying capacity for the young pigeons
K2 = 2 # carrying capacity for the nest predators
K3 = 3 # carrying capacity for the raptors (Quanto maior for a capacidade, maior será a amplitude do predador. Se a capacidade for maior, o predador chega a tal capacidade e se mantém constante)
## Allee effect: fenômeno biológico que relaciona a densidade populacional com o bem-estar individual de uma espécie
K4 = 5 ## campacidade do recurso
e = 1 # Allee parameter for the adult pigeons / Ainda estamos trabalhando mais ainda para entender melhor o efeito do parametro
q = F0/A0 #quota de recursos por individuo
k = Fmax/2   #constante de consumo do recurso (considerando o recurso que a ararinha azul mais come)
# Quantidade maxima em que o recurso começa a diminuir
r =  1 #taxa de crescimento do recurso (considerando semente de fruto)
out = odeint(model_2, y0, t, args=(a, m, p1, p2, l1, l2, b1, b2,b3, g1, g2, ps1, ps2, K1, K2, K3, K4, e, q, k, r))

plt.plot(t, out[:, 0], 'g', label='Ararinhas Jovens')
plt.plot(t, out[:, 1], 'b', label='Ararinhas Adultos')
plt.plot(t, out[:, 2], 'r', label='Predadores de Ninho')
plt.plot(t, out[:, 3], 'y', label='Aves de Rapina')
plt.plot(t, out[:, 4], 'cyan', label='Alimento')
plt.legend(loc='best')
plt.title('Interações entre Ararinhas-Azuis: Predadores e Fonte de Alimento')
plt.xlabel('Tempo (meses)')
plt.ylabel('Taxas populacionais')
plt.show()
plt.show()
```

* Os parâmetros $A$, $Y$, $N$ e $R$ representam, respectivamente, os valores iniciais para Ararinhas-Azuis adultas, Ararinhas-Azuis jovens,  predadores de ninho e aves de rapinas. Os valores aplicados em um ambiente real pode variar consideravelmente quando analisado quantitativos de predadores de ninhos e aves de rapinas. Os valores foram inicialmente escolhidos de modo a se aproximarem mais harmoniosamente da representação exposta no artigo estudado.

* A maturação $μ$ foi baseada na estimativa para Araras-Azuis [8], meramente ilustrativa, lembrar que a limitação de dados corroboram em um certo distanciamento de uma análise mais promissora do modelo.

* A taxa de mortalidade de $π_i$ de Ararinhas-Azuis adultas e jovens são definidas sobre a densidade populacional influenciadas por requisitos naturais como habitat, disponibilidade de recursos, reprodução, dentre outros. A dedução não é trivial e a sua complexidade depende de muitos parâmetros, muitas vezes de uma precisão estatísitca robusta. Estamos definindo emppiricamente tal dado, de modo a se aproximar do comportamento esperado pelo gráfico.

* A determinação da taxa de crescimento dos predadores, tal como abordado em relação à taxa de mortalidade das Ararinhas, não emerge de forma meramente descritiva. Portanto, torna-se imperativa a realização de uma análise de natureza mais substancial e robusta, a fim de compreender profundamente esses aspectos. Estamos definindo empiricamente tal dado, de modo a se aproximar do comportamento esperado pelo gráfico.

* O handling time é o tempo necessário para lidar com a presa de modo que outra presa não possa ser atacada simultaneamente. Pode incluir perseguir, subjugar, comer, digerir. A exatidão do parametro está sendo precisamente estudado, dado a disponibilidade de recursos, para assim aumentar a acurácia do modelo. Estamos definindo empiricamente tal dado, de modo a se aproximar do comportamento esperado pelo gráfico.

* O modelo de Holling tipo 3, uma ferramenta analítica de grande relevância, desempenha um papel crucial ao diferenciar entre diversos tipos de interações ecológicas, tais como dependência de recursos, dependência de consumidores e dependência de proporção. Este modelo também proporciona uma abordagem para avaliar o impacto da densidade crescente de consumidores sobre a taxa per capita de consumo. Inicialmente, temos $N$, representando a densidade do recurso no início do experimento. A variável $f$ denota a taxa de consumo instantâneo, isto é, o número de presas consumidas por um consumidor em uma unidade de tempo. Além disso, o parâmetro $a^2$ corresponde à taxa de ataque, caracterizando o ritmo no qual os consumidores se engajam na busca e captura de presas. O último elemento, $h$, alude ao tempo de manuseio, indicando o tempo que um consumidor despende para efetuar o consumo de uma presa após capturá-la.

    O cálculo é definido utilizando utiliando handlling time, o parámetro de capacidade da populção analisada (beneficiada), e o indice de ataque (taxa de predatismo), o qual será utilado a sua raiz (represnetado por "a"). [3] [8]. O resultado foi baseado nos parametros previamente explicados.

   $𝒇(N) = \dfrac{aN^2}{1+ahN^2} $

* Os parâmetros de predação $γ_i$ determinam a influência que a presença da população de Ararinhas-Azuis exerce na taxa de crescimento das populações de predadores. A definição para esse parâmetro não é atingida por métodos triviais e sua determinação exata necessitaria de um estudo de campo dedicado e sistemático. Estamos definindo empiricamente tal dado, de modo a se aproximar do comportamento esperado pelo gráfico [6].

* Os parâmetros de capacidade de carga $K_i$ definem a capacidade populacional máxima de uma espécie biológica que um ambiente pode sustentar indefinidamente considerando a disponibilidade de recursos, como alimento, habitat, água, entre outras necessidades fisiológicas. Os parâmetros são determinados com orientação à capacidade populacional limítrofe de indivíduos intencionada para as testagens do modelo.

* O parâmetro $ϵ$ representa o impacto do efeito Allee na população de Ararinhas-Azuis livres na natureza. O efeito Allee foi descrito pela primeira vez em 1930 por Warder Clyde Allee, por meio de análises experimentais de grupos de peixinhos dourados em um tanque. Trata-se de um fenômeno biológico que correlaciona a densidade populacional de uma espécie com o bem-estar individual da população. Ele determina que a agregação pode aumentar as taxas de sobrevivência dos indivíduos e que cooperação pode ser crucial para a evolução geral de uma estrutura social. Como é possível notar, a determinação precisa do parâmetro é muito complexa e depende de uma observação mais próxima do comportamento das populações selvagens, como por exemplo, o mecanismo mais comumente observado é a limitação do acasalamento, que causa efeitos Allee tanto em animais quanto em plantas, assim como omo a tendência dos predadores de ficarem saciados e de parar de aumentar o consumo, sendo este último muito dependente do tipo de predador, e não da preza. Com base nisso, dado ao nível de complexidade e depêndencia realista dos dados e comportamento entre predadores, ambiente e comportamento do animal, determinamos valores arbitrários para o coeficiente.

 O efeito Alle forte ocorre quando a população possui um limiar de crescimento, onde acima dele a população cresce e abaixo dele a população vai à extinção
 Pór outro lado, o fraco, ocorre quando a população não possui tal limiar, contudo, possui um fator de crescimento pequeno mas ainda existe uma influência positiva na taxa de crescimento populacional à medida que a densidade aumenta, isso devido a vários fatores, como cooperação entre indivíduos ou proteção contra predadores.



### 3.3. Formulação do modelo de ODE idealizado





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def model_2(y, t, a, u, p1, p2, l1, l2, b1, b2, y1, y2, w1, w2, K1, K2, K3, e):
  Y, A, N, R = y
  sol_system = [a*A*(1-(A/K1))*((A/e)-1) - u*Y - N*(b1*Y**2/(1+b1*w1*Y**2)) - p1*Y,
                u*Y - R*(b2*A**2/(1+b2*w2*A**2)) - p2*A,
                l1*N*(1-(N/K2)) + y1*Y*N,
                l2*R*(1-(R/K3)) + y2*R*A]
  return sol_system
```

```python

#------- condições iniciais de mesma quantidade de inserção para as populações--------

#------- parametros iniciais --------

Y0 = 2 # starting young pigeons
A0 = 2 # starting adult pigeons
N0 = 2 # starting nest predators
R0 = 2 # starting raptors (predators of adult pigeons)
y0 = Y0, A0, N0, R0
t = np.linspace(0, 25, 150) #análise do experiemento nos primeiros anos de reinserção (consideraremos o modelo em meses)

#------- constantes --------

a = 0.08 # birth of new pigeons / definido a partir de dado particular do comportamento de uma comunidade de 52 Ararinhas em cativeiro
m = 6 # maturation of young pigeons into adult pigeons (tempo de virarem adultos eh 6 meses)
p1 = 0.30 # death of young pigeons (Dado empírico, com valor mais alto considerando a fragilidade de filhotes de aves)
p2 = 0.15 # death of adult pigeons (Dado empírico, capacidade de morte de um adulto em porcentagem)
l1 = 0.09 # growth of nest predators (Dado empírico, consideramos baixo dado que o índice de nascimento de filhotes é menor do que a de total de adultos Ararinha)
l2 = 0.1 # growth of raptors
## Holling Type parameters: taxa populacional do predador em função da densidade da presa.
## Type I: taxa de crescimento linear; assume que o consumo e digestão do alimento não interferem na sua busca.
## Type II: taxa de crescimento hiperbólica (aR/(1+ahR); assume que a busca por alimento é limitada pelo seu processamento.
## Type III: taxa de crescimento hiperbólica potencial (aR^k/(1+ahR^k)); simula processo de aprendizado do predador.
b1 = 0.07 # Holling Type parameter for the young pigeons/nest predators system
b2 = 0.096 # Holling Type parameter for the adult pigeons/raptors system
g1 = 0.01 # predation for the nest predators (Dado empirico,cada predado come 1)
g2 = 0.01 # predation for the raptors (1 a cada)
ps1 = 4 # handling time for the young pigeons (Dado empírico, estimativa de tempo (dado em horas))
ps2 = 4 # handling time for the adult pigeons (Dado empírico, estimativa de tempo (dado em horas))
K1 = 40 # carrying capacity for the young pigeons
K2 = 2 # carrying capacity for the nest predators
K3 = 3 # carrying capacity for the raptors (Quanto maior for a capacidade, maior será a amplitude do predador. Se a capacidade for maior, o predador chega a tal capacidade e se mantém constante)
## Allee effect: fenômeno biológico que relaciona a densidade populacional com o bem-estar individual de uma espécie
e = 1 # Allee parameter for the adult pigeons / Ainda estamos trabalhando mais ainda para entender melhor o efeito do parametro

out = odeint(model_2, y0, t, args=(a, m, p1, p2, l1, l2, b1, b2, g1, g2, ps1, ps2, K1, K2, K3, e))

plt.plot(t, out[:, 0], 'g', label='Ararinhas Jovens')
plt.plot(t, out[:, 1], 'b', label='Ararinhas Adultos')
plt.plot(t, out[:, 2], 'r', label='Predadores de Ninho')
plt.plot(t, out[:, 3], 'y', label='Aves de Rapina')
plt.legend(loc='best')
plt.title('Relacao Populacional Ararinhas-Azuis x Predadores')
plt.xlabel('Tempo (meses)')
plt.ylabel('Taxas populacionais')
plt.show()
plt.show()
```
